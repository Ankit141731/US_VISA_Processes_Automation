```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings

warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df = pd.read_csv("EasyVisa.csv")
df.head(5)
```

```python
numeric_value_features = [feature for feature in df.columns if df[feature].dtype != "O"]
categorical_features = [feature for feature in df.columns if df[feature].dtype == "O"]

print(f"we have {len(numeric_value_features)} numerical features : {numeric_value_features}")
print(f"We have {len(categorical_features)} Categorical features : {categorical_features}")
```

```python
for column in categorical_features :
    print(df[column].value_counts(normalize = True)*100)
    print("-------------------")
```

```python
plt.figure(figsize=(15,10))
plt.suptitle("Univariate Analysis of Numerical Features", fontsize = 20 , fontweight = "bold"  )

i = 1
for column in numeric_value_features:
    plt.subplot(2,2,i)
    i = i + 1
    sns.kdeplot(x = df[column] ,color = "blue" )
    plt.xlabel(column)
    
```

```python
plt.figure(figsize = (25 , 15))

plt.suptitle("Univariate Analysis of Categorical Features" , fontsize = 20 , fontweight = "bold")

i = 1
for column in categorical_features:

    plt.subplot(3 , 3 , i)
    sns.countplot(x = df[column] , color = "blue")
    plt.xlabel(column)
    i = i+1
```

```python
discrete_features = [feature for feature in numeric_value_features if len(df[feature].unique()) <= 25]
continious_features = [feature for feature in numeric_value_features if len(df[feature].unique()) > 25]

print(f'We have {len(discrete_features)} discrete features : {discrete_features}')
print(f'\nWe have {len(continious_features)} continuous_features : {continious_features}')
```

```python
df[numeric_value_features].corr()
```

```python
from scipy.stats import chi2_contingency
chi2_test = []
for feature in categorical_features:
    if chi2_contingency(pd.crosstab(df['case_status'], df[feature]))[1] < 0.05:
        chi2_test.append('Reject Null Hypothesis')
    else:
        chi2_test.append('Fail to Reject Null Hypothesis')
result = pd.DataFrame(data=[categorical_features, chi2_test]).T
result.columns = ['Column', 'Hypothesis Result']
result
```

```python
continues_features=[feature for feature in numeric_value_features if len(df[feature].unique())>=10]
print('Num of continues features :',continues_features)
```

```python
clr1 = ['#1E90FF', '#DC143C']
fig, ax = plt.subplots(3, 2, figsize=(10,12))
fig.suptitle('Distribution of Numerical Features By Case Status', color='#3C3744',
             fontsize=20, fontweight='bold', ha='center')
for i, col in enumerate(continues_features):   
    sns.boxplot(data=df, x='case_status', y=col,  palette=clr1, ax=ax[i,0])
    ax[i,0].set_title(f'Boxplot of {col}', fontsize=12)
    sns.histplot(data=df, x=col, hue='case_status', bins=20, kde=True, 
                 multiple='stack', palette=clr1, ax=ax[i,1])
    ax[i,1].set_title(f'Histogram of {col}', fontsize=14)
fig.tight_layout()
fig.subplots_adjust(top=0.90)
```
